```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud  # Pastikan ini ada

# Membaca dataset df_laptop
df_laptop = pd.read_csv("laptops.csv")

# Menghitung jumlah kemunculan setiap merk laptop
laptop_counts = df_laptop['Laptop'].value_counts()  # Menggunakan kolom 'Laptop' untuk merk

# Membuat string dengan merk laptop dan frekuensinya
laptop_text = ' '.join([f'{merk} ' * count for merk, count in laptop_counts.items()])

# Membuat word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(laptop_text)

# Menampilkan word cloud
plt.figure(figsize=(12, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  # Menghilangkan axis
plt.title('Word Cloud Merk Laptop', fontsize=16)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

df_laptop = pd.read_csv("laptops.csv")

x = df_laptop[['CPU', 'RAM', 'Storage']]
y = df_laptop['Final Price']

label_encoder = LabelEncoder()
x['CPU'] = label_encoder.fit_transform(x['CPU'])

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=4)

model = LinearRegression()

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

plt.figure(figsize=(10, 6))

plt.scatter(X_test.iloc[:, 0], y_test, label='Harga Aktual', color='blue')
plt.scatter(X_test.iloc[:, 0], y_pred, label='Harga Prediksi', color='red')

plt.title('Prediksi Harga Laptop berdasarkan CPU', fontsize=16)
plt.xlabel('CPU (encoded)', fontsize=12)
plt.ylabel('Final Price', fontsize=12)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
cpu_input = 'Intel Core i7'
ram_input = 16
storage_input = 512
cpu_encoded = label_encoder.transform([cpu_input])[0]

input_data = [[cpu_encoded, ram_input, storage_input]]

prediksi_harga = model.predict(input_data)

print(f"Prediksi harga laptop dengan CPU {cpu_input}, RAM {ram_input}GB, dan Storage {storage_input}GB adalah: {prediksi_harga[0]:,.2f}")
```

```python
cpu_input = 'Intel Core i7'  # Ganti dengan inputan CPU lain jika diperlukan
ram_input = 16  # Ganti dengan inputan RAM (dalam GB)
storage_input = 512  # Ganti dengan inputan Storage (dalam GB)

# Mengubah nilai input CPU menjadi nilai numerik (encoded)
cpu_encoded = label_encoder.transform([cpu_input])[0]

# Mengubah inputan menjadi integer
ram_input = int(ram_input)
storage_input = int(storage_input)

# Membuat input data untuk prediksi
input_data = [[cpu_encoded, ram_input, storage_input]]

# Melakukan prediksi menggunakan model
prediksi_harga = model.predict(input_data)

# Menampilkan hasil prediksi harga
print(f"Prediksi harga laptop dengan CPU {cpu_input}, RAM {ram_input}GB, dan Storage {storage_input}GB adalah: {prediksi_harga[0]:,.2f}")
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# Melakukan prediksi pada data pengujian (X_test)
y_pred = model.predict(X_test)

# Menghitung MAE, MSE, dan RMSE
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

# Menampilkan hasil evaluasi
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
```

Hasil evaluasi model regresi linear menggunakan MAE, MSE, dan RMSE memberikan gambaran akurasi model dalam memprediksi harga laptop:

- MAE (Mean Absolute Error) menunjukkan rata-rata selisih antara prediksi dan nilai sebenarnya. Semakin kecil MAE, semakin akurat prediksi model.
- MSE (Mean Squared Error) mengukur rata-rata kuadrat kesalahan. MSE yang lebih kecil berarti kesalahan prediksi lebih terkendali.
- RMSE (Root Mean Squared Error) adalah akar kuadrat MSE, mengukur kesalahan rata-rata dalam satuan yang sama dengan data asli. Nilai RMSE yang kecil menunjukkan prediksi yang lebih akurat.

Secara keseluruhan, semakin kecil ketiga nilai ini, semakin baik model dalam memprediksi harga laptop.
